# Plotting Graphs

## Importing Libraries

```python
import matplotlib.pyplot as plt
import numpy as np
```

## Taking Inputs

```python
xlim = int(input("Enter x limit :"))
ylim = int(input("Enter y limit :"))
size = int(input("Enter Size : "))
x = np.array([np.random.randint(0,xlim) for i in range(size)])
y = np.array([np.random.randint(0,xlim) for i in range(size)])

print(x)
print(y)
```

## Line Plot

```python
plt.plot(x,y)
plt.xlabel("X axis Label")
plt.ylabel("Y axis Label")
plt.show()
```

## Bar Graph

```python
plt.bar(x,y)
plt.xlabel("X axis Label")
plt.ylabel("Y axis Label")
plt.show()
```

```python
## Histogram
```

```python
plt.hist(x)
plt.xlabel("X axis Label")
plt.ylabel("Y axis Label")
plt.show()
```

## Pie Chart Exploded

```python
plt.pie(x,explode=[np.random.random() for i in range(size)],shadow=True)
plt.show()
```

## Pie Chart Simple

```python
plt.pie(x,shadow=True)
plt.show()
```
